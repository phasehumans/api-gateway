```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/ChaiCode Full Stack Data Science Cohort/Phase 3 - Machine Learning/RandomForest_3July_11AM/WineQT.csv')
data.head()
```

```python
data.drop(columns=['Id'], axis=1, inplace=True)
```

```python
data.columns
```

```python
X = data.drop(columns=['quality'], axis=1)
y = data['quality']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=44)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train, y_train)
```

```python
y_pred = dt_clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import tree
plt.figure(figsize=(20, 20))
tree.plot_tree(dt_clf, filled=True)
```

```python
sample_data = data.head(20)
X_sample = sample_data.drop(columns=['quality'], axis=1)
y_sample = sample_data['quality']
```

```python
sample_data
```

```python
from collections import Counter
Counter(y_sample)
```

```python
X_sample.columns
```

```python
sample_model = DecisionTreeClassifier()
sample_model.fit(X_sample, y_sample)
```

```python
from sklearn import tree
plt.figure(figsize=(20, 20))

tree.plot_tree(sample_model, filled=True)
plt.savefig('sample_model_tree.png')
plt.show()
```

```python
sample_model_entropy = DecisionTreeClassifier(criterion='entropy')
sample_model_entropy.fit(X_sample, y_sample)
```

```python
from sklearn import tree
plt.figure(figsize=(20, 20))

tree.plot_tree(sample_model_entropy, filled=True)
plt.savefig('sample_model_tree_entropy.png')
plt.show()
```

```python
importances = sample_model_entropy.feature_importances_
print(importances)
```

```python
feature_importance_df = pd.DataFrame({
    'feature': X_sample.columns,
    'feature_importance': importances
}).sort_values(by='feature_importance', ascending=False)
```

```python
feature_importance_df
```
