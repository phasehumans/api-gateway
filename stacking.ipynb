Import all the required libraries

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import StackingClassifier
from sklearn.model_selection import train_test_split
```

Load the dataset

```python
X, y = load_breast_cancer(return_X_y=True)
```

```python
X
```

```python
y
```

Data Split into train and test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=44)
```

Define the base learners

```python
base_learners = [
    ('dt', DecisionTreeClassifier(random_state=44)),
    ('lr', LogisticRegression(random_state=44)),
    ('knn', KNeighborsClassifier())
]
```

Define the meta learner

```python
meta_learner = DecisionTreeClassifier(random_state=44)
```

Model Training: Build and Fit the Stacking Classifier

```python
stack_clf = StackingClassifier(
    estimators = base_learners,
    final_estimator = meta_learner
)

stack_clf.fit(X_train, y_train)
```

Model Prediction

```python
y_pred = stack_clf.predict(X_test)
y_pred
```

Model Evaluation

```python
accuracy_score(y_test, y_pred)
```

Adaboost Implementation

```python
from sklearn.ensemble import AdaBoostClassifier
ada_model = AdaBoostClassifier(random_state=44)

ada_model.fit(X_train, y_train)
y_pred = ada_model.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
