```python
from nltk.util import ngrams

# Example sentence
sentence = "<s> Books open the door to knowledge. </s>"

print(f"Given Sentence: {sentence}")

# Tokenize the sentence into words
words = sentence.split()

# Create unigrams from the list of words
unigrams = ngrams(words, 1)
print(" ")
print("Unigrams in given sentence are:")
print(" ")

# Print the unigrams
for unigram in unigrams:
    print(unigram)

# Create bigrams from the list of words
bigrams = ngrams(words, 2)
print(" ")
print("Bigrams in given sentence are:")
print(" ")

# Print the bigrams
for bigram in bigrams:
    print(bigram)

# Create trigrams from the list of words
trigrams = ngrams(words, 3)
print(" ")
print("Trigrams in given sentence are:")
print(" ")

# Print the trigrams
for trigram in trigrams:
    print(trigram)
```
