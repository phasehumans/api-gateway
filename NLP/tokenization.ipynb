```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
text= "The children are playing on the basketball court in the evening"
```

```python
tokens= word_tokenize(text.lower())
print("Tokenization", tokens)
```

```python
stop_words= set(stopwords.words('english'))
print(stop_words)
print(len(stop_words))
```

```python
# filtered_tokens= [word for word in tokens if word not in stop_words]
# print("stop word remove", filtered_tokens)
```

```python
filter_token= []
remove_words= []

for i in tokens:
    if i not in stop_words:
        filter_token.append(i)
    else:
        remove_words.append(i)

print(filter_token)
print(remove_words)
```

```python
stemmer= PorterStemmer()
# print(stemmer)

print(stemmer.stem('evening'))

stemmed_tokens= []
for i in filter_token:
  stemmed_tokens.append(stemmer.stem(i))

print(stemmed_tokens)
```

```python
lemmatizer= WordNetLemmatizer()

print(lemmatizer.lemmatize('evening'))

lemmatized_tokens= []
for i in filter_token:
  lemmatized_tokens.append(lemmatizer.lemmatize(i))

print(lemmatized_tokens)
```
