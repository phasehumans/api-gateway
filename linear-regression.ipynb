Import all the required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

Data Preparation

```python
dataset = pd.DataFrame(housing.data, columns=housing.feature_names)
dataset
```

```python
dataset.head()
```

Task 1: Perform the EDA on the above dataset and provide the quality data by tomorrow.

```python
dataset['Price'] = housing.target
dataset.head()
```

Data Split

```python
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X
```

```python
y
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

Data Standardization

Task 2: Why we are not doing fit_transform in the test set and let's say if we do fit_transform in the test set what will be the impact on the model's result?

**Data Leakage (major problem in the industry work)**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

Model Training

Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

Lasso Regression (L1 Regularization)

```python
from sklearn.linear_model import Lasso
model_lasso = Lasso()
```

Ridge Regression (L2 Regularization)

```python
from sklearn.linear_model import Ridge
model_ridge = Ridge()
```

Elastic Net Regression (Combination of L1 and L2)

```python
from sklearn.linear_model import ElasticNet
model_elastic_net = ElasticNet()
```

```python
model_lasso.fit(X_train, y_train)
model_ridge.fit(X_train, y_train)
```

```python
model_elastic_net.fit(X_train, y_train)
```

Model Prediction

```python
y_cap = model.predict(X_test)
print(y_cap)
```

```python
y_cap_lasso = model_lasso.predict(X_test)
print(y_cap_lasso)
```

```python
y_cap_ridge = model_ridge.predict(X_test)
print(y_cap_ridge)
```

```python
y_cap_elasticNet = model_elastic_net.predict(X_test)
print(y_cap_elasticNet)
```

Model Evaluation

```python
residuals = y_test - y_cap
residuals
```

```python
# Normality -> Distribution plot of the above residuals == it should be normally distributed
sns.distplot(residuals, kde=True)
```

Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print(mean_squared_error(y_test, y_cap))
print(mean_absolute_error(y_test, y_cap))
print(np.sqrt(mean_squared_error(y_test, y_cap)))
```

R-squared vs Adjusted R-squared

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_cap)
print(r2)
```
